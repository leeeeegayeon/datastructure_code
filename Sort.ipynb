```python
//bubble sort
#include<iostream>
#include<vector>
#include<algorithm>
using namespace std;

int main() {
	ios::sync_with_stdio(false);
	cin.tie(nullptr);
	int N;
	cin >> N;
	vector<pair<int, int>> A(N);
	for (int i = 0;i < N;i++) {
		cin >> A[i].first;
		A[i].second = i;
	}
	sort(A.begin(), A.end());
	int Max = 0;
	for (int i = 0;i < N;i++) {
		Max = max(Max, A[i].second - i);
	}
	cout << Max + 1 << endl;
	return 0;

}
```

```python
//selection sort
#include<iostream>
#include<vector>
#include<string>
using namespace std;

int main() {
	ios::sync_with_stdio(false);
	cin.tie(nullptr);
	string s;
	cin >> s;
	vector<int> A(s.size(), 0);
	for (int i = 0; i < s.size();i++) {
		A[i] = stoi(s.substr(i, 1)); //각각의 원소를 정수로 저장
		cout << A[i] << " ";
	}
	for (int i = 0;i < s.size();i++) {
		int max = i; //i번째 원소를 최대값으로 가정
		for (int j = i + 1;j < s.size();j++) {
			if (A[j] > A[max]) max = j; //최대값 갱신
		}
		if (A[i] < A[max]) { //최대값과 i번째 원소 교체
			int temp = A[i]; //임시저장
			A[i] = A[max]; //i번째 원소에 최대값 저장
			A[max] = temp; //최대값 위치에 i번째 원소 저장
		}
	}
	cout << "\n";
	for (int i = 0; i < A.size();i++) {
		cout << A[i] << " ";
	}
}
```

```python
//insertion sort
#include<iostream>
#include<vector>
using namespace std;

int main() {
	ios::sync_with_stdio(false);
	cin.tie(nullptr);
	int N;
	cin >> N;
	vector<int> A(N, 0);
	vector<int> S(N, 0); //누적합 저장할 배열
	for(int i=0;i<N;i++)
		cin >> A[i];
	for (int i = 1;i < N;i++) {
		int insertPoint = i; //삽입 위치
		int insertValue = A[i]; //삽입할 값
		for (int j = i - 1;j >= 0;j--) { //현재 범위에서 삽입 위치 찾기
			if (A[j] < A[i]) {
				insertPoint = j + 1; //삽입 위치 찾음
				break;
			}
			if (j == 0) //맨 앞까지 왔을 때
				insertPoint = 0; //삽입 위치는 맨 앞
		}
		for (int j = i;j > insertPoint;j--) //삽입 위치부터 한 칸씩 뒤로 밀기
			A[j] = A[j - 1]; //한 칸씩 뒤로 밀기
		A[insertPoint] = insertValue;//삽입
	}
	S[0] = A[0]; //누적합 초기화
	for (int i = 1;i < N;i++)
		S[i] = S[i - 1] + A[i]; //누적합 계산
	int sum = 0;
	for(int i=0;i<N;i++)
		sum += S[i]; //누적합의 누적합 계산
	cout << sum << "\n";
}
```
